```python
import pandas as pd
```

```python
df = pd.read_csv('dataset.csv')
df.info()
```

```python
df['HomeTeam'].value_counts()
```

```python
# Convert the 'date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Extract the year, month, and weekday
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['weekday'] = df['Date'].dt.day_name()  # This will give the weekday as string

df.head(20)
```

```python
# Convert 'date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Filter the DataFrame for rows starting from '2024-07-26'
filtered_df = df[df['Date'] >= '2024-07-26']

# Count occurrences of each team in 'HomeTeam' from the filtered data
team_counts = filtered_df['HomeTeam'].value_counts()

# Display the result
print("Number of occurrences for each team from 2024-07-26:")
print(team_counts)
```

```python


# Define the date range
start_date = '2023-07-28'
end_date = '2024-07-26'

# Filter the DataFrame for the date range (excluding 2024-07-26)
filtered_df = df[(df['Date'] >= start_date) & (df['Date'] < end_date)]

# Count how many unique teams are in 'HomeTeam' within this date range
team_count = filtered_df['HomeTeam'].nunique()

# Display the result
print(f"Number of unique teams from {start_date} to {end_date} (excluding {end_date}): {team_count}")
```

```python

# Group by 'HomeTeam' and get the list of unique 'year' values for each team
team_years = df.groupby('HomeTeam')['year'].apply(lambda x: list(x.unique()))
# Display the result
print("Unique 'year' values for each team:")
print(team_years)
```

```python
# Convert 'time' column to datetime
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.time

# Function to categorize the time
def categorize_time(t):
    if t >= pd.to_datetime('12:00:00').time() and t < pd.to_datetime('13:30:00').time():
        return 'noon'
    elif t >= pd.to_datetime('13:30:00').time() and t < pd.to_datetime('17:00:00').time():
        return 'afternoon'
    elif t >= pd.to_datetime('17:00:00').time() and t < pd.to_datetime('19:30:00').time():
        return 'evening'
    elif t >= pd.to_datetime('19:30:00').time() or t < pd.to_datetime('06:00:00').time():
        return 'night'
    else:
        return 'morning'

# Apply the function to the 'time' column
df['time_of_day'] = df['Time'].apply(categorize_time)
df.head()
```
