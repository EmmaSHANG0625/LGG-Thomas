#Check Price by using interquartile range (IQR) method to remove outliers

```python
# Calculate the quartiles and the interquartile range (IQR)
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

# Define the outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out the outliers
df = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
```

```python
#covert 'PEB_grade' back to 'PEB'
PEB = {
   9: 'A++',
 8: 'A+',
 7: 'A',
 6: 'B',
 5: 'C',
 4: 'D',
 3: 'E',
 2: 'F',
 1: 'G'
}
df['PEB'] =df['PEB_grade'].map(PEB)
df.to_csv("Data for Machine Learning.csv", index = True)
```

```python
df.info()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
#creat OneHot vector
df_PEB = df[["PEB"]]
encoder = OneHotEncoder()
df_PEB_1hot = encoder.fit_transform(df_PEB)
# 将稀疏矩阵转换为密集数组
df_PEB_1hot_array = df_PEB_1hot.toarray()

```

```python
# 获取独热编码后的列名
encoded_columns = encoder.get_feature_names_out(['PEB'])

# 将独热编码数组转换为 DataFrame
df_PEB_1hot_df = pd.DataFrame(df_PEB_1hot_array, columns=encoded_columns)

# 将原始 DataFrame 与独热编码后的 DataFrame 合并
df = pd.concat([df, df_PEB_1hot_df], axis=1)
```

```python
df.to_csv("Data for Machine Learning.csv", index = True)
```

```python
df.info()
```
