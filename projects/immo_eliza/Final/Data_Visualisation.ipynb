```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('cleaned_data.csv')
df_appartments = pd.read_csv('data_appartment.csv')
df_houses = pd.read_csv('data_house.csv')
print(len(df))
```

colaration of each data

```python

appartments_counts = df_appartments['province'].value_counts()
houses_counts = df_houses['province'].value_counts()
#print(appartments_counts, houses_counts)
```

```python
#combine into a Dataframe
df_properties_distribution = pd.DataFrame({'Houses': houses_counts, 'Appartments': appartments_counts})
#properties_distribution_df = properties_distribution_df.reset_index.rename(columns={'index': 'province'})
print(df_properties_distribution)
```

```python
#plot a bar chart
df_properties_distribution_sorted = df_properties_distribution.sort_values(by='Houses', ascending=False)
df_properties_distribution_sorted.plot(kind='bar', figsize=(10, 6))

plt.xlabel('Provinces')
plt.ylabel('Properties Amount')
plt.title('Properties Distribution')
plt.xticks(rotation=45)
plt.legend()

plt.show()
```

```python
# Group the data by province and calculate the mean value of price per sqm
mean_pricesqm_appartments = df_appartments.groupby('province')['price_per_sqm'].mean()
mean_pricesqm_houses = df_houses.groupby('province')['price_per_sqm'].mean()

# Step 3: Display the result
print(mean_pricesqm_appartments, mean_pricesqm_houses)
```

```python
df_price_distribution = pd.DataFrame({'Houses': mean_pricesqm_houses, 'Appartments': mean_pricesqm_appartments})
df_price_distribution_sorted = df_price_distribution.sort_values(by='Houses', ascending=False)
df_price_distribution_sorted.plot(kind='bar', figsize=(10, 6))

plt.xlabel('Provinces')
plt.ylabel('priceSqm (eruo/m2)')
plt.title('PriceSqm Distribution')
plt.xticks(rotation=45)
plt.legend()

plt.show()
```
