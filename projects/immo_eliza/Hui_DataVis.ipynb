```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Hui.csv')
```

```python
#Highest pricesqm
top5_pricesqm = df['price_per_sqm'].nlargest(3)
top5_pricesqm_rows = df.nlargest(3, 'price_per_sqm')
print(top5_pricesqm_rows) 
top5_pricesqm_rows['Url'].tolist()
```

```python

plt.figure(figsize=(10, 6))
plt.scatter(df['construction_year'], df['price_per_sqm'], alpha=0.5)

# Adding labels and title
plt.xlabel('Construction Year')
plt.ylabel('PriceSqm (eruo/m2)')
plt.title('Distribution of Price by Construction Year')

#zoom y
plt.ylim(1725, 2050)
plt.ylim(-1000, 50000)

# Customizing the ticks and grid
plt.xticks()
plt.grid(True)

# Display the plot
#plt.tight_layout()
plt.show()
```
